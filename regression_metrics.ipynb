# MEAN ABSOLUTE ERROR

Mean Absolute Error (MAE) is a commonly used regression evaluation metric that measures the average magnitude of errors between predicted values and actual values, without considering their direction (positive or negative).


========================================================

MAE tells you on average how much your predictions are wrong, in the same units as the target variable.

üìå Example:
If MAE = 3, it means your predictions are off by 3 units on average.




============================================================


Advantages of MAE

Easy to understand

Error is in the same unit as the output (e.g., marks, price, temperature).

Less sensitive to outliers (than MSE)

Errors are not squared, so extreme values don‚Äôt dominate the result.

Simple and intuitive

Straightforward interpretation for beginners and practitioners.

Robust for real-world noisy data

Works well when you don‚Äôt want to heavily penalize large errors.

‚ùå Disadvantages of MAE

Does not penalize large errors strongly

A big mistake and a small mistake contribute linearly.

Sometimes large errors should matter more.

Not differentiable at zero

Absolute value causes issues for gradient-based optimization (minor but relevant in theory).

Less sensitive to variance

Two models with very different error distributions can have the same MAE.

Not ideal when large errors are critical

For cases like medical diagnosis or finance, bigger errors may need higher punishment.

MAE	‚ùå Low	‚úÖ Yes
MSE	‚úÖ High	‚ùå No
RMSE	‚úÖ High	‚úÖ Yes

When interpretability matters

When outliers should not dominate

When you want a simple and robust metric

Mean Squared Error (MSE) is one of the most widely used regression evaluation metrics in machine learning and statistics. It measures the average of the squared differences between predicted values and actual values.
-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
‚úÖ Advantages of MSE

Strongly penalizes large errors

Squaring magnifies big mistakes ‚Üí useful when large errors are costly.

Mathematically convenient

Differentiable everywhere ‚Üí works well with gradient descent.

Widely used and well-understood

Standard loss function for Linear Regression.

Sensitive to variance in errors

Helps distinguish models with similar average errors but different spreads.


------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


‚ùå Disadvantages of MSE

Highly sensitive to outliers

A single extreme error can dominate the entire metric.

Unit is squared

If target is in meters, MSE is in square meters ‚Üí less interpretable.

Over-penalizes large errors

Sometimes unrealistic in noisy real-world data.

Not robust

Performs poorly when data contains many outliers.

üß† Simple Explanation

Numerator ‚Üí error made by your model

Denominator ‚Üí error made if you always predict the mean



üìå Example:

R¬≤ = 0.80 ‚Üí Model explains 80% of variance

R¬≤ = 0.00 ‚Üí Model is no better than predicting the mean

R¬≤ = -0.50 ‚Üí Model is worse than mean prediction ‚ùå
```

R¬≤ = 1.0 ‚Üí The model is perfect and explains 100% of the variance in the target variable.

R¬≤ between 0.7 and 0.9 ‚Üí The model is very good and explains most of the data variation.

R¬≤ between 0.3 and 0.7 ‚Üí The model has moderate performance; it explains some variance but still has noticeable errors.

R¬≤ = 0.0 ‚Üí The model performs no better than predicting the mean of the target variable.

R¬≤ < 0 ‚Üí The model is very poor and performs worse than the mean predictor.




```
‚úÖ Advantages of R¬≤ Score

Easy to interpret

Expressed as a percentage of explained variance.

Scale-independent

Units cancel out ‚Üí works across different datasets.

Good for model comparison

Higher R¬≤ generally indicates a better fit.

Widely used & accepted

Standard metric in regression problems.
```


```
‚ùå Disadvantages of R¬≤ Score

Does not indicate prediction accuracy directly

High R¬≤ ‚â† low MAE/MSE always.

Increases with more features

Even useless features can increase R¬≤ ‚Üí misleading.

Not reliable for non-linear relationships

Can be low even if predictions are good.

Sensitive to outliers

Outliers can inflate or deflate R¬≤.

üß† Simple Explanation

R¬≤ always increases when you add more features (even useless ones ‚ùå)

Adjusted R¬≤ checks whether the new feature is really helpful

If a feature is irrelevant, Adjusted R¬≤ decreases

üìå So, it gives a more honest measure of model performance.

```



```
Interpretation (Normal Format)

Adjusted R¬≤ = 1.0 ‚Üí Perfect model

High Adjusted R¬≤ ‚Üí Model explains data well using relevant features

Adjusted R¬≤ ‚âà R¬≤ ‚Üí Added features are useful

Adjusted R¬≤ < R¬≤ ‚Üí Model has unnecessary features

Adjusted R¬≤ < 0 ‚Üí Very poor model

‚úÖ Advantages of Adjusted R¬≤

Penalizes extra features

Prevents overfitting.

Better for model comparison

Especially when models have different numbers of features.

More reliable than R¬≤

Reflects true explanatory power.

‚ùå Disadvantages of Adjusted R¬≤

Only works for linear regression

Not suitable for non-linear models.

Less intuitive than R¬≤

Slightly harder to explain to beginners.

Still sensitive to outliers

Inherits limitations of R¬≤.

üìå When to Use Adjusted R¬≤?

When comparing multiple regression models

When feature count varies

Along with MAE / MSE / RMSE for complete evaluation
